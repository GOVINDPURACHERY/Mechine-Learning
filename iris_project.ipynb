```python
#Visualisation
```

```python
sns.pairplot(iris_toy,hue='species', height=3, aspect=1)
```

```python
iris_toy.hist(edgecolor='black',linewidth = 1.2, figsize=(12,8))
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.violinplot(x='species', y='sepal_length',data=iris_toy)
plt.subplot(2,2,2)
sns.violinplot(x='species', y='sepal_width', data=iris_toy)
plt.subplot(2,2,3)
sns.violinplot(x='species', y='petal_length', data=iris_toy)
plt.subplot(2,2,4)
sns.violinplot(x='species', y='petal_width', data=iris_toy)
```

```python
iris_toy.boxplot(by='species', figsize=(12,8));
```

```python
pd.plotting.scatter_matrix(iris_toy, figsize=(12,10))
plt.show()
```
